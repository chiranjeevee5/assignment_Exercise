```python
#Declare the libraries

import pandas as pd
```

```python
#Read the Csv file

dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#Converting numbers to String for Nominal Dataset

dataset=pd.get_dummies(dataset)
```

```python
dataset
```

```python
dataset.columns
```

```python
#Input Dataset

independent=dataset[['age', 'bmi', 'children', 'sex_female', 'sex_male',
       'smoker_no', 'smoker_yes']]
```

```python
independent
```

```python
#Output Dataset

dependent=dataset[["charges"]]
```

```python
dependent
```

```python
# Using sklearn class, Split the Train and Test data values. 
# Test_size and random_state are few of the value to be declared in the function.

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(criterion='mse',n_estimators=100, random_state=0,max_features='log2')
regressor.fit(X_train,Y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
```
