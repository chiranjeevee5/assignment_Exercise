```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3: 5].values
# X = dataset.iloc[1:5, 3: 5].values
# X = dataset.iloc[:, 3: 5]
```

```python
X
```

```python
from sklearn.cluster import Birch
brc = Birch(n_clusters = None)
y_brc = brc.fit_predict(X)
```

```python
y_brc
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised['Cluster_group']=y_brc
```

```python
supervised
```

```python
supervised.to_csv("cluster_brc.csv",index=False)
```
